```python
"""professor/Osama shalabia
  assistant professor/Ehab Ahmed
  student/barbara karam"""

#shortening the name of neccessary libs for convenience
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# defining the func
def func(x, y):
    return np.exp(-(x**2 + y**2))

# Define the func grad
def grad_func(x, y):
    return -2*x*np.exp(-(x**2 + y**2)), -2*y*np.exp(-(x**2 + y**2))

# Create a grid since the task is supposed to graphed
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

# Calculate the function and the grad for the plot points
Z = func(X, Y)
Gx, Gy = grad_func(X, Y)

# Create a 3D plot of the function
fig = plt.figure(figsize=(18, 6))

ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis')
ax1.set_title('the function')

# Create a 2D contour plot of the gradient
ax2 = fig.add_subplot(133)
G = np.sqrt(Gx**2 + Gy**2) #to make the color of the point show its magnitude
ax2.quiver(X[::5], Y[::5], Gx[::5], Gy[::5] ,G[::5], cmap='viridis')
ax2.set_title('contour of the gradient')

plt.show()
```
