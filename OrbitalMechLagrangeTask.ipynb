```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

line='*'*50

# User inputs(mu,r0,v0,delta theta)
#mu
central_body = input("Is the central body Earth? (yes/no): ")
if central_body.lower() == 'yes':
    mu = 398600
else:
    mass=float(input("Please enter the mass of the central body in kilograms:"))
    mu = 6.67430e-11 * mass
#r0
r0 = np.array([float(input("Enter r0 x-component: ")),
               float(input("Enter r0 y-component: ")),
               float(input("Enter r0 z-component: "))])
#v0
v0 = np.array([float(input("Enter v0 x-component: ")),
               float(input("Enter v0 y-component: ")),
               float(input("Enter v0 z-component: "))])
#delta theta
delta_theta = np.radians(float(input("Enter delta theta: ")))

# Calculate the magnitude of r0 and v0
r0_mag = np.linalg.norm(r0)
v0_mag = np.linalg.norm(v0)

# Calculate the radial velocity
vr0 = np.dot(v0, r0) / r0_mag

#Calculate the angular momentum
h=np.cross(r0,v0)
h_mag=np.linalg.norm(h)

#calculate the r magnitude

#to simplify the expression
a=(h_mag**2)/mu
b=(a/r0_mag)-1
c=(h_mag*vr0)/mu

r_mag=a/(1+(b*np.cos(delta_theta))-(c*np.sin(delta_theta)))

# Calculate the Lagrange coefficients
f = 1- (((mu*r_mag)/(h_mag**2))*(1-np.cos(delta_theta)))
g = ((r_mag*r0_mag)/h_mag)*np.sin(delta_theta)

#to simplify the expression
d=(mu*r0_mag)/(h_mag**2)
E= 1-np.cos(delta_theta)

g_dot = 1- (d*E)
f_dot = (1/g)*((f*g_dot)-1)

# Calculate the final velocity vector
r = (f*r0) + (g*v0)
v = (f_dot*r0) + (g_dot*v0)

#showing the steps to find the error in the input in case there's any
print('the lagrange coeffiecients are:')
print('f =',f)
print('g = ',g)
print('gdot = ',g_dot)
print('fdot = ',f_dot)

print(line)

print("position vector (r): ", r)
print("velocity vector (v): ", v)

print(line)


# Initialize arrays to store the position and velocity at each step
r_values = np.zeros((360, 3))
v_values = np.zeros((360, 3))

# Set the initial values
r_values[0] = r0
v_values[0] = v0

# Loop over the time steps
for delta_theta in range(1, 360):
    #update lagrange coeffiecients
    r_mag=a/(1+(b*np.cos(delta_theta))-(c*np.sin(delta_theta)))

    # Calculate the Lagrange coefficients
    f = 1- (((mu*r_mag)/(h_mag**2))*(1-np.cos(delta_theta)))
    g = ((r_mag*r0_mag)/h_mag)*np.sin(delta_theta)

    #to simplify the expression
    d=(mu*r0_mag)/(h_mag**2)
    E= 1-np.cos(delta_theta)

    g_dot = 1- (d*E)
    f_dot = (1/g)*((f*g_dot)-1)
    # Update the position and velocity
    r = (f*r0) + (g*v0)
    v = (f_dot*r0) + (g_dot*v0)
    # Store the new values
    r_values[delta_theta] = r
    v_values[delta_theta] = v

fig = plt.figure()

# Add a 3D subplot for the position
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot(r_values[:, 0], r_values[:, 1], r_values[:, 2])
ax1.set_title('Position vs. Time')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')

# Add a 3D subplot for the velocity
ax1 = fig.add_subplot(122, projection='3d')
ax1.plot(v_values[:, 0], v_values[:, 1], v_values[:, 2])
ax1.set_title('velocity vs. Time')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')

# Display the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

mu = 398600
# r0
r0 = np.array([3500,
               4800,
               2558])
# v0
v0 = np.array([-2,
               3,
               7])

# Create a DataFrame to store the results
result_df = pd.DataFrame(columns=['Theta', 'f', 'g', 'f_dot', 'g_dot', 'r', 'v'])

for theta in range(1, 361):
    # Convert theta to radians
    delta_theta = np.radians(theta)

    # Calculate the magnitude of r0 and v0
    r0_mag = np.linalg.norm(r0)
    v0_mag = np.linalg.norm(v0)

    # Calculate the radial velocity
    vr0 = np.dot(v0, r0) / r0_mag

    # Calculate the angular momentum
    h = np.cross(r0, v0)
    h_mag = np.linalg.norm(h)

    # Calculate the r magnitude
    a = (h_mag**2) / mu
    b = (a/r0_mag) - 1
    c = (h_mag*vr0) / mu

    r_mag = a / (1 + (b*np.cos(delta_theta)) - (c*np.sin(delta_theta)))

    # Calculate the Lagrange coefficients
    f = 1 - (((mu*r_mag)/(h_mag**2))*(1-np.cos(delta_theta)))
    g = ((r_mag*r0_mag)/h_mag)*np.sin(delta_theta)

    # Simplify the expression
    d = (mu*r0_mag)/(h_mag**2)
    E = 1 - np.cos(delta_theta)

    g_dot = 1 - (d*E)
    f_dot = (1/g)*((f*g_dot)-1)

    # Calculate the final velocity vector
    r = (f*r0) + (g*v0)
    v = (f_dot*r0) + (g_dot*v0)

    # Append the results to the DataFrame
    result_df = result_df.append({
        'Theta': theta,
        'f': f,
        'g': g,
        'f_dot': f_dot,
        'g_dot': g_dot,
        'r': r,
        'v': v
    }, ignore_index=True)

# Save the DataFrame to an Excel file
result_df.to_excel('output.xlsx', index=False)
```
